```python
# this is used to emulate display device on headless hosts like google collab. Uncomment if yoiu run this code in such
"""
import os
os.environ["SDL_VIDEODRIVER"] = "dummy"
"""

# define required libraries and import them

import pygame
import pygame.surfarray as surfarray
import matplotlib.pyplot as plt
import random

# initialize pygame library
pygame.init()
```

```python
# Create a pygame screen to draw shapes
screen = pygame.display.set_mode((1000, 1000))
screen.fill((1*random.randint(0,55), 1*random.randint(0,5), 1*random.randint(0,5)))

# used to update the entire display surface on the screen
screen.blit(screen, (0, 0))
```

```python
# randomly define colors
colors = [(1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255)), (1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255)), (1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255)), (1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255)), (1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255)), (1*random.randint(0,255), 1*random.randint(0,255), 1*random.randint(0,255))]
```

```python
# set number of shapes to draw
count = 400

# randomly define shapes and their characteristics
shapesArray = []
for i in range(count):
    # define center of the object
    x = random.randint(0, 1000)
    y = random.randint(0, 1000)

    # define radius for circle shape
    radius = random.randint(8, 40)

    # randomly define color
    color = random.choice(colors)

    # randomly choose border size
    border = random.randint(0, 5)

    # randomly choose shape
    shape = random.choice([0,1,2,3])

    # add created shape to shapesArray
    shapesArray.append((x, y, radius, color, border, shape))
```

```python
# Draw shapes in the array
for i in shapesArray:
    if i[5] == 0:
        # draw circles
        pygame.draw.circle(screen, i[3], (i[0], i[1]), i[2], i[4])
    elif i[5] == 1:
        # draw ellipses
        pygame.draw.ellipse(screen, i[3], (i[0], i[1], 40, 25), i[4])
    elif i[5] == 2:
        # draw rectangles
        pygame.draw.rect(screen, i[3], (i[0], i[1], 20, 20), i[4])
    elif i[5] == 3:
        #  draw arcs
        pygame.draw.arc(screen,i[3],(i[0], i[1], 40, 25), 60, 90, i[4])
```

```python
# display changes made to the screen
pygame.display.flip()

# adding pygame surface array into matplotlib so that it is able to input into matplotlib
pygame.surfarray.use_arraytype("numpy")
image = surfarray.array3d(screen)
```

```python
# Display the image using Matplotlib
plt.imshow(image)
plt.axis("off")
plt.show()

# Quit Pygame
pygame.quit()
```
